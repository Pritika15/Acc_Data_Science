```python
#1-Read Dataset using cuDF
filename='netflix_titles.csv'
df=cudf.read_csv(filename)
df
```

```python
#2-creates a pandas dataframe
import pandas as pd
pdf=pd.read_csv('netflix_titles.csv')
pdf
```

```python
#3-creates cudf dataframe from pandas dataframe
pdf2cudf=cudf.from_pandas(pdf)
#4-display df
pdf2cudf
```

```python
#7-Drop Na Value
df=df.dropna()
```

```python
#10-Sort the dataframe according to the year the record was released (latest first), Refer to sort_values function, which takes the target column name and the sorting mode.
df.sort_values('release_year',ascending=False)
```

```python
#11-Find the number of movies and shows that are available using value_counts
df.value_counts(['type'])
```
